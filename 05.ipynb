```python
class Gestor:
    def __init__(self):
        self.condiciones_iniciales()
    
    def condiciones_iniciales(self):
        self._ingresos = 0
        self._egresos = dict(
            medicos=0,
            domesticos=0,
            ocio=0,
            educacion=0
        )
        self._ahorros = 0
        self._gasto_total = 0
    
    def ingresar_ingresos(self):
        while True:
            try:
                self._ingresos = int(input("Registre sus ingresos: "))
                print()
            except:
                print("\nEntrada no válida")
            else: break
    
    def ingresar_gastos(self):
        while True:
            self.mostrar_menu_gastos()
            while True:
                try:
                    opcion = int(input("Selecione una opcion: "))
                    print()
                except:
                    print("\nEntrada no válida")
                else:
                    if opcion < 1 or opcion > 6:
                        print("Valores fuera del rango")
                        print("Ingrese valores correctos")
                    else: break
            if opcion == 6:
                print()
                break
            while True:
                try:
                    gasto = int(input("Registre sus gastos: "))
                    print()
                except:
                    print("\nEntrada no válida")
                else: break
            if opcion == 1:
                self._egresos["medicos"] += gasto
            elif opcion == 2:
                self._egresos["domesticos"] += gasto
            elif opcion == 3:
                self._egresos["ocio"] += gasto
            elif opcion == 4:
                self._ahorros += gasto
            elif opcion == 5:
                self._egresos["educacion"] += gasto
            print()
    
    def listar_gastos(self):
        print("Listado de gastos según tipo:")
        for tipo, gasto in self._egresos.items():
            print(f"{tipo}: ${gasto}")
        print()
    
    def obtener_gastos_totales(self):
        self._gasto_total = self._ahorros
        for tipo, gasto in self._egresos.items():
            self._gasto_total += gasto
        print(f"El gasto total es de ${self._gasto_total}")
        print()
    
    def analizar_finanzas(self):
        if self._ingresos ==  self._gasto_total:
            print(f"Usted ha gastado más en el área de {max(self._egresos, key=self._egresos.get)}")
            print("CONSEJOS ...")
        elif self._ingresos ==  self._gasto_total:
            print("FELICITACIONES ...")
        else:
            print("Usted gasta más de lo que gana")
            print("CONSEJOS PARA MEJORAR SU SALUD FINANCIERA ...")
        print()
    
    def mostrar_menu_gastos(self):
        print("Seleccione el tipo de gasto que desea registrar:")
        print("----------------------")
        print("1. Gastos médicos")
        print("2. Gastos domésticos")
        print("3. Ocio")
        print("4. Ahorros")
        print("5. Educación")
        print("6. Salir de registro de gastos\n")
        
```

```python
gestor = Gestor()
```

```python
def main():
    print("Bienvenido a nuestro gestor financiero\n")
    gestor.ingresar_ingresos()
    gestor.ingresar_gastos()
    gestor.listar_gastos()
    gestor.obtener_gastos_totales()
    gestor.analizar_finanzas()
    gestor.condiciones_iniciales()
```

```python
if __name__ == "__main__":
    main()
```
